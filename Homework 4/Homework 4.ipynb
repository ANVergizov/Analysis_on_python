#### 2. На основе переменной quality с оценками вина создать новую переменную good: если quality > 5, то 1, иначе 0.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data

df = pd.read_csv('winequality-red.csv', sep = ';')
df.head(7)
```

```python
df.info()
```

```python
#Create a new column

df['good'] = df['quality'].apply(lambda x: 1 if x > 5 else 0)
df.head()
```

### 3. Исследовать, какие химические характеристики вина влияют на то,
###    окажется оно хорошим или плохим


Начну исследование с построения матрицы корреляций, с применением библиотеки Seaborn

Что бы посмотреть, что в большей или меньшей мере влияет на качество вина и его оценку 

```python
# Create a correlation matrix

correlation = df.corr()
correlation
```

##### Думаю можно не рассматривать признаки с корреляцией в диапазоне от 0,2 до -0,2

```python
# Create a heatmap

correlation[np.abs(correlation) < 0.2] = 0

plt.figure(figsize=(12, 10))
sns.heatmap(correlation, cmap = 'ocean', annot = True)
plt.title('Correlation matrix')
plt.show()
```

#### Вывод по матрице корреляций
На матрице видно что явно влияют на переменную "good" только 5 признаков, но так как рассматриваемая переменная построена на основе признака "quality" нам нельзя сделать валидных выводов относительно этой взаимосвязи.
Далее посмотрим поближе на взаимосвязи между остальными 4 признаками:

#### 'alcohol' 'sulphates' 'volatile acidity' 'total sulfur dioxide'

### Alcohol

Характеристика показывающая самую высокую корреляцию

Построим для него диаграмму рассеивания и посмотрим на нее

```python
# Prepare data to vizualization

alcohol = df.groupby('alcohol')['good'].sum().reset_index()
mean_alco = df['alcohol'].median()



mean_alco, alcohol.head()
```

```python
# Create a scattering diagram

plt.style.use('tableau-colorblind10')

plt.figure(figsize=(10,7))

x = np.linspace(0, 50, 2)

plt.scatter(alcohol['good'], alcohol['alcohol'])
plt.plot(x, [median_alco]*2, label='median', c='r')
plt.xlabel('Количество положительных оценок качества')
plt.ylabel('Содержание алкоголя')
plt.title('Зависимость оценки красного вина от содержания объемной доли спирта', size=17 )
plt.legend(fontsize=12)
plt.show()
```

Можем наблюдать, что вина с высокой объемной долей спирта гораздо резже получают хорошие оценки, на графике можем наблюдать обратную регрессию, можно сказать, что большее количество положительных оценок получают вина в диапазоне от 9 до 12 % объемной доли спирта

```python
df.groupby('alcohol')['good'].agg(
    count='sum').reset_index().sort_values(by='count', ascending=False).head(10)
```

## Sulphates

Попробуем рассмотреть содержание сульфатов к оценке и качеству вина на гистограмме

```python
df['sulphates'].value_counts()
```

```python
#Prepare data

sulphates = df.groupby('sulphates')['good'].sum().reset_index()
sulphates.head()
```

```python

plt.figure(figsize=(10,7))

plt.plot(sulphates['sulphates'],sulphates['good'])
plt.title('Влияние содержания сульфатов на оценку вина', size=17)
plt.xlabel('Содержание сульфатов')
plt.ylabel('Количество положительных оценок')

plt.show()
```

На графике видно, что наилучшие оценки получили вина с содержанием сульфатов от 0.50 до 0.75, можно выделить этот диапазон как максимально благоприятный для получения оценки выше 5 

## volatile acidity

Посмотрим на распределение признака летучей кислотности и подготовим данные

```python
df['volatile acidity']
```

```python
volatile = df.groupby('volatile acidity')['good'].sum().reset_index()
volatile
```

Построим гистограмму для определения взаимосвящи между кислотностью и оценками вина

```python
plt.figure(figsize=(10,8))

x = np.arange(0, 1.5, 0.1)

plt.bar(volatile['volatile acidity'],volatile['good'])
plt.xticks(x)
plt.title('Влияние летучей кислотности на оценку вина', size=17)
plt.xlabel('Летучая кислотность')
plt.ylabel('Количество положительных оценок')

plt.show()
```

Можно сказать наверняка, что вина с кислотностю выше 1.0 получают положительные оценки реже чем остальные

## total sulfur dioxide

Рассмотрим последний признак и построим график который отраз ит связь между этим показателем и оценкой

```python
df['total sulfur dioxide'].max()
```

```python
tsd = pd.crosstab(df['total sulfur dioxide'],['good']).reset_index().sort_values(by='good',ascending=False)
tsd.head()
```

```python
plt.figure(figsize=(10, 7))

plt.bar(tsd['total sulfur dioxide'], tsd['good'])
plt.title('Влияние диоксида серы на оценку вина', size=17)
plt.xlabel('Количество диоксида серы')
plt.ylabel('Количество положительных оценок')

plt.show()
```
